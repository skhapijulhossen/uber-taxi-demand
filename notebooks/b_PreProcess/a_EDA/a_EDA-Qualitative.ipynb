## Time Series Analysis

### Trend

```python
df.date
```

```python
df.date = pd.to_datetime(df.date, format='%d-%m-%y')
df.groupby('date')[['Net_Price']].mean()
```

```python
# Plot the responses for different events and regions
plt.figure(figsize=(20,10))
sns.lineplot(
    data=df.groupby('date')[['Net_Price']].mean(),
    x="date", y="Net_Price", legend='auto'
)
plt.show()
```
